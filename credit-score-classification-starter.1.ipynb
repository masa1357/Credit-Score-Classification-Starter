# グリッドサーチ
min_samples_splitとmin_samples_leafをサーチ

```python
X_train = train_df.drop(['Poor_Credit_Score'], axis=1).to_numpy()
y_train = train_df['Poor_Credit_Score'].to_numpy()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# グリッドサーチのパラメータ範囲を定義
param_grid = {
    'min_samples_split': [2, 10, 20, 30, 40],
    'min_samples_leaf': [1, 2, 5, 10, 15],
    'max_depth':[5,6,7,8,9]
}

# モデルの定義
dt = DecisionTreeClassifier()

# グリッドサーチの定義
grid_search = GridSearchCV(dt, param_grid, cv=5)  # 5-fold cross-validation

# グリッドサーチの実行
grid_search.fit(X_train, y_train)

# 最適なパラメータの表示
print('Best parameters:', grid_search.best_params_)

# 最適なパラメータで訓練されたモデルの取得
best_dt = grid_search.best_estimator_
```

```python
from sklearn.model_selection import KFold, cross_val_score
kfold = KFold(n_splits=10)
scores = cross_val_score(best_dt, X_train, y_train, cv=kfold)
print("K-Fold CV Score: ", scores.mean())
```
